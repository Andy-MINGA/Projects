```python
# Now, we create a scatterplot matrix showing pairwise scatterplots between
# the first three variables in our dataframe.

import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix

# Select the columns of interest
subset = college[['Top10perc', 'Apps', 'Enroll']]

# Create the scatterplot matrix
scatter_matrix(subset, figsize=(8, 8), alpha=0.7, diagonal='hist')

# Show the plot
plt.show()
```

```python
# Now, we create a side-by-side boxplot of Outstate vs Private
college.boxplot(column='Outstate', by='Private', figsize=(8, 6))

# Add title and labels
plt.title('Out-of-State Tuition by Private/Public')
plt.suptitle('')  # Removes the default automatic title
plt.xlabel('Private')
plt.ylabel('Outstate Tuition ($)')
plt.show()
```

```python
# Now, we create a new qualitative variable, called Elite, by binning the
# Top10perc variable into two groups based on whether or not the
# proportion of students coming from the top 10% of their high
# school classes exceeds 50%.

college['Elite'] = pd.cut(college['Top10perc'],
[0,50,100],
labels=['No', 'Yes'])
```

```python
# Now, we use the .value_counts() method on the Elite column

elite_counts = college['Elite'].value_counts()
print(elite_counts)
```

```python
# Now, we use the boxplot() method to produce side-by-side boxplots of Outstate versus Elite.

college.boxplot(column='Outstate', by='Elite', figsize=(8, 6))

plt.title('Out-of-State Tuition by Elite Status')
plt.suptitle('')  # Remove the automatic pandas title
plt.xlabel('Elite')
plt.ylabel('Outstate Tuition ($)')
plt.show()
```

```python
# We now plot 4 histograms in a 2x2 grid with different bin sizes

# Create a 2x2 grid of subplots
fig, axes = plt.subplots(2, 2, figsize=(12, 10))  # 2 rows, 2 columns

# Flatten axes array for easy iteration
axes = axes.flatten()

# List of columns to plot
columns = ['Apps', 'Accept', 'Enroll', 'Outstate']

# Different bin sizes for each plot
bins_list = [10, 20, 30, 40]

# Loop through columns, axes and bins simultaneously
for ax, col, bins in zip(axes, columns, bins_list):
    college[col].plot.hist(bins=bins, ax=ax, alpha=0.7, color='skyblue')
    ax.set_title(f'Histogram of {col} (bins={bins})')
    ax.set_xlabel(col)
    ax.set_ylabel('Frequency')

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plots
plt.show()
```
